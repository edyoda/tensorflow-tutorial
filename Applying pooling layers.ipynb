* A pooling layer method is to reduce the number of trainable parameters in a smart way
* average pooling - specific blocks avg is extracted  
* max pooling - max of specific block is extracted
* By reducing the trainable params, we reduce training time & avoid overfitting

```python
import numpy as np

from keras.utils import np_utils
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Flatten
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers import Conv2D, MaxPooling2D
from keras.optimizers import Adam

from keras.callbacks import EarlyStopping
```

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
img_rows, img_cols = X_train[0].shape[0], X_train[0].shape[1]
X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)
```

```python
X_train = X_train.astype('float32')/255.
X_test = X_test.astype('float32')/255.
```

```python
n_classes = len(set(y_train))
y_train = np_utils.to_categorical(y_train, n_classes)
y_test = np_utils.to_categorical(y_test, n_classes)
```

```python
y_train.shape
```

```python
model = Sequential()
```

```python
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
```

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Conv2D(256, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
batch_size = 128
n_epochs = 200

callbacks = [EarlyStopping(monitor='val_acc', patience=5)]
```

```python
model.fit(X_train, y_train, batch_size=batch_size, epochs=n_epochs, verbose=1, validation_split=0.2, callbacks=callbacks)
```
