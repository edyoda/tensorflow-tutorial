```python
import pandas as pd
```

```python
network_data = pd.read_csv('Data/UNSW_NB15_training-set.csv',index_col='id')
```

```python
network_data.head()
```

```python
network_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_cols = ['proto','service','state']
```

```python
for col in cat_cols:
    le = LabelEncoder()
    network_data[col] = le.fit_transform(network_data[col]) 
```

```python
network_data.head()
```

```python
network_data.info()
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
select = SelectKBest(k=25)
```

```python
network_tf = select.fit_transform(network_data.drop(['attack_cat'],axis=1),network_data.label)
```

```python
network_tf[:5]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model_relu = Sequential()
model_relu.add(Dense(100, input_dim=25, activation='sigmoid')) 
# model_relu.add(Dense(50, activation='relu'))
# model_relu.add(Dense(25, activation='relu'))
model_relu.add(Dense(2, activation='softmax'))
```

```python
model_relu.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
model_relu.summary()
```

```python
from keras.utils import to_categorical
```

```python
target = to_categorical(network_data.label)
```

```python
target
```

```python
model_relu.fit(network_tf, target,epochs=100,batch_size=20000)
```
