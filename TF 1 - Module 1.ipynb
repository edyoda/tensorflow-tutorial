### Hello World in TensorFlow
* h, w & hw are nodes of graph.
* run() will execute the graph

```python
h = tf.constant("Hello")
```

```python
w = tf.constant("Zeke")
```

```python
hw = h + w
```

```python
hw
```

```python
with tf.Session() as sess:
    res = sess.run(hw)
    tf.summary.FileWriter('tensorboard/new-logs',sess.graph)
```

### Fetches 

```python
a = tf.constant(5)
b = tf.constant(7)
c = tf.constant(10)
d = tf.multiply(a,b)
e = tf.add(b,c)
f = tf.subtract(d,e)
with tf.Session() as sess:
    out = sess.run(f)
```

```python
with tf.Session() as sess:
    fetches = [a,b,c,d,e,f]
    outs = sess.run(fetches)

print(outs)
```

```python
A = tf.constant( [
                   [1,2,3],
                   [4,5,6]
                 ])
```

```python
b = tf.constant( [1,0,1])
```

```python
b = tf.expand_dims(b,1)
```

```python
b.shape
```

```python
A.shape
```

```python
matmul = tf.matmul(A,b)
```

```python
with tf.Session() as sess:
    o = sess.run(matmul)
    print (sess.run(b))
```

```python
o
```

```python
with tf.Session() as sess:
    fetches = [A,b]
    outs = sess.run(fetches)
```

```python
outs
```

```python
r = tf.transpose(A)
```

```python
with tf.Session() as sess:
    fetches = [r]
    outs = sess.run(fetches)
```

```python
outs
```

### Variables & Placeholders

```python
init_vals = tf.random_normal((1,5),0,1)
```

```python
init_vals.shape
```

```python
var = tf.Variable(init_vals)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    print ( sess.run(var))
```

### Placeholders 

```python
ph = tf.placeholder(tf.int16)
```

```python
d = tf.multiply(ph,2)
    
```

```python
with tf.Session() as sess:
    print(sess.run(d, feed_dict={ph:5}))
    print(sess.run(d, feed_dict={ph:15}))
```
